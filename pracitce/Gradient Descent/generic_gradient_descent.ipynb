```python
def mx(m,x):
    new_m = m*x
    sum=0
    for i in range(14):
        sum +=new_m[i]
    return sum
```

```python
def step(points,learning_rate,m):
    m_slope = np.zeros([14])
    n = len(points)
    for i in range(n):
        y = points[i,13]
        x = points[i,:13]
        x = np.append(x,[1])#for the intercept "c"
        #calculating for m1,m2,m3...
        for j in range(13):
            m_slope[j] +=(-2/n)*(y-mx(m,x))*points[i,j]
        #for intercept "c"
        m_slope[13] += (-2/n)* (y -mx(m,x))
    new_m = m - learning_rate * m_slope
    print(new_m)
    return new_m
    
```

```python
def gd(points, learning_rate,iterations):
    m=np.zeros([14])#13 for m and 1 more for c
    for i in range(iterations):
        m = step(points,learning_rate,m)
    return m
```

```python
def run():
    data= np.loadtxt("train.csv",delimiter = ",")
    learning_rate=0.001
    iterations=100
    m = gd(data,learning_rate,iterations)
    print(m)
    
```

```python
run()
```
